```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# # of hours a AI student spends on their courses in a month
np.random.seed(10)
no_hours = np.random.normal(40.0, 10.0, 10000)

plt.hist(no_hours, 50)
plt.show()
```

## Task 1:

```python
#Calculating Mean, Standard Deviation, and Variance
mean_hours = np.mean(no_hours)
std_hours = np.std(no_hours)
var_hours = np.var(no_hours)
```

```python
# Reporting the results
print(f"Mean: {mean_hours:.2f} hours")
print(f"Standard Deviation: {std_hours:.2f} hours")
print(f"Variance: {var_hours:.2f} hours^2")
```

## Task 2:



```python
#Creating the function:
def create_array(num_elements, mean, std):
    return np.random.normal(mean, std, num_elements)

def print_array_info(arr):
    print(f'Mean: {np.mean(arr)}')
    print(f'Std :  {np.std(arr)}')
    print(f'Var :  {np.var(arr)}')
    #plot the array
    plot_array(arr)

    print('=====================')

def plot_array(arr):
    plt.hist(arr, 50)
    plt.show()
```

```python
#Arrays with varying standard deviation
original_arr = create_array(10000, 40.0, 10.0)
arr_1 = create_array(10000, 40.0, 30.0)
arr_2 = create_array(10000, 40.0, 50.0)

print('Original array (10.000 elements, mean 40.0, std 10.0):')
print_array_info(original_arr)

print('Array 1 (10.000 elements, mean 40.0, std 30.0):')
print_array_info(arr_1)

print('Array 2: (10.000 elements, mean 40.0, std 50.0):')
print_array_info(arr_2)
```

```python
#Arrays with varying mean
arr_3 = create_array(10000, 60.0, 10.0)
arr_4 = create_array(10000, 80.0, 10.0)

print('Original array (10.000 elements, mean 40.0, std 10.0):')
print_array_info(original_arr)

print('Array 3 (10.000 elements, mean 60.0, std 10.0):')
print_array_info(arr_3)

print('Array 4: (10.000 elements, mean 80.0, std 10.0):')
print_array_info(arr_4)
```

## Conclusion:
Change in std: When the standard deviation increases, the spread of values in the array also increases. A smaller standard deviation results in a more concentrated distribution, while a larger standard deviation leads to a more dispersed distribution.

Change in mean: Changing the mean of the arrays without altering the standard deviation results in a shift in the central tendency of the distributions. The consistent standard deviation ensures that the spread of values around the mean remains constant, highlighting the independent impact of the mean on the central location of the data.
